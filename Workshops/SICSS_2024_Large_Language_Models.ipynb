**Prompt memory** in models like ChatGPT refers to the ability of the AI to retain context from earlier interactions within a single session. This feature enhances the model's responses by allowing it to reference and build upon previous inputs, making conversations more coherent and contextually relevant.

In the context of models like ChatGPT, **temperature** controls the randomness of text generation. A low temperature (e.g., 0.1) produces more predictable and straightforward responses, whereas a high temperature (e.g., 1.0) makes the model more likely to use less probable words, leading to more creative, diverse, or nuanced outputs. This feature allows users to adjust the AI's output to match the desired level of creativity or precision for their specific task.
